```python
# imports
import numpy as np
from skimage.io import imread
```

```python
# load image
im = imread("images/bumblebee.png")
```

```python
from py_scripts.utils import make_compatible_with_torch

im_torch = make_compatible_with_torch(im)
```

```python
from py_scripts.visual import visualize_tensor

visualize_tensor(im_torch)
```

```python
# build conv2d filter
import torch.nn as NN

KERNEL_SIZE = 2

filter = NN.Conv2d(
    in_channels=3,
    out_channels=3,
    kernel_size=KERNEL_SIZE,
    stride=2,
    bias=False,
    padding=0
)
```

```python
filter.weight.shape
```

```python
visualize_tensor(filter(im_torch))
```

```python
import torch

w = torch.empty(3, 3, KERNEL_SIZE, KERNEL_SIZE)

w[:, 0, :, :] = 1/4
w[:, 1, :, :] = 1/4
w[:, 2, :, :] = 1/4
w
```

```python
w.shape
```

```python
filter.weight = NN.Parameter(w)
```

```python
filter_image = filter(im_torch)
visualize_tensor(filter_image)
```

```python
filter_image.shape
```

```python
avg_pool = NN.AvgPool2d(kernel_size=2, stride=(1, 1))

visualize_tensor(avg_pool(im_torch))
```
