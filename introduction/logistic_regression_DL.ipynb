```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the data (cat/non-cat)
from py_scritps.utils import load_dataset

train_set_x_orig, train_set_y, test_set_x_orig, test_set_y, classes = load_dataset()
```

```python
def show_img(index: int, data_img = train_set_x_orig, labels_img = train_set_y):
    # observation
    img = data_img[index]
    label = np.squeeze(labels_img)[index]

    plt.title(f"{classes[label].decode('utf-8')}")
    plt.imshow(img)
    return
```

```python
show_img(56)
```

```python
# implementation of sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```
